# Data Visualization:

# Using Matplotlib and Seaborn Libraries:

```python
df.Team.value_counts()
```

```python
sns.set(rc={'figure.figsize':(20,7)})
a = sns.countplot(x = 'Team', data = df)
for i in a.containers:
    a.bar_label(i)
```

# Observations:
    
1. More number of Top Batsman are from England Team. 15 Batsman out of 100 top batsman are from England Team.
2. Secondly, 13 Batsman are from India Team who are in the Top 100 Batsman list.
3. Third, Pakistan Batsman from Pakistan Team are 12 in number in Top 100 Batsman.
4. Australia and Sri Lanka Batsman are in same number that is 11 in Top 100 Batsman List.
5. Afganistan Team Batsman are Least in number as compared to other team batsman in 100 Top Batsman list.

```python
df.groupby(['Team'], as_index = False)['Century'].sum().sort_values(by = 'Century', ascending = False)
```

```python
sns.set(rc={'figure.figsize':(30,15)})
#with plt.rc_context(dict(sns.axes_style("whitegrid"),
# **sns.plotting_context(font_scale=30))):
sns.set_style("whitegrid")
sns.set(font_scale = 2)   
a = sns.countplot(x = 'Team', data = df, hue = 'Century', palette = 'cool')
plt.legend(ncol=4)
a.set(ylim=(0, 6));
```

```python
sns.violinplot(x='Team',y='Century',data=df.sort_values('Team'));
```
